## Phase transition in the 1nn with antiferromagnetic interaction and small longitudinal field

#### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Data

```python
hmaxs=[2.5,2.6,np.e,2.8,2.9]
ls=[16,32,64,128]

x={}
z={}
xx={}
v={}

for l in ls:
    for h in hmaxs:
        data=np.load(f'data/1nn_xx_z_x/dataset_241122/test_periodic_{l}_l_{h}_h_100_n.npz')
        x[(l,h)]=data['magnetization_x']
        z[(l,h)]=data['density']
        v[(l,h)]=data['potential']
        xx[(l,h)]=data['correlation']
        
```

#### Analysis of the magnetization x

```python
m={}

for l in ls:
    stg=(-1)**np.arange(l)
    for h in hmaxs:
        m[(l,h)]=np.average(np.abs(x[(l,h)]*stg[None,:]))
        # plt.plot(x[(l,h)][0],label=f'l={l}, h={h}')
        # plt.plot(v[(l,h)][0])
        # plt.legend()
        # plt.show()
        
    plt.plot(hmaxs,[m[(l,h)] for h in hmaxs],label=f'l={l}')
plt.legend()
plt.show()
```

#### Correlation analysis

```python
l=128
for h in hmaxs:
    plt.plot(np.average(np.abs(xx[(l,h)][:,0]),axis=0))
plt.loglog()
plt.show()
    
```
